RESULTS CALC

```python
input = open("input.txt", 'r')

ctr = 0
correct = 0
for l in input:
    if ctr%2==0:
        if l[:5] != "Done.": correct+=1
    ctr+=1

print("correct: "+str(correct))

input_file = open("input.txt", 'r')

total_inference_time = 0.0
counter = 0

for line in input_file:
    if "Inference," in line:
        before_inference = line.split("Inference,")[0]
        try:
            start_index = before_inference.rfind("(") + 1
            end_index = before_inference.find("ms", start_index)
            pre_inference_time = float(before_inference[start_index:end_index].strip())
            total_inference_time += pre_inference_time
            counter += 1
        except (ValueError, IndexError):
            continue

print(f"Total Pre-Inference Time: {total_inference_time} ms")
print(f"Total Numbers Extracted: {counter}")

print(f"Avg Inference time: {total_inference_time/counter}")
```
